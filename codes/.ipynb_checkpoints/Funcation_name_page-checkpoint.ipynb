```python
import numpy as np
import pandas as pd
import json
```

```python
def convert_to_HTML_format(function_names_list, webpage_name):
    html_string = '''
   
        <html>
        <head>
            <link rel="stylesheet" href="https://maxcdn.bootstrapcdn.com/bootstrap/3.3.1/css/bootstrap.min.css">
            <style>body{ margin:0 100; background:whitesmoke; }</style>
        </head>
        <body>

            <h1>Function names</h1>
            
            <!-- *** Function Lists *** - -->
            
            <table class="table table-striped">
                <thead>
                    <tr style="text-align: right;">
                    <th>GO Label</th>
                    </tr>
                </thead>
                <tbody>
                    <script>
                        var functionNames = ''' + function_names_list + '''
                        console.log(functionNames);
                        
                        // Get the table body element
                        var tbody = document.querySelector("tbody");
                        
                        
                        // Generate the table rows dynamically
                        functionNames.forEach(function (name) {

                            // Create a new row
                            var row = document.createElement("tr");

                            // Create a new cell
                            var cell = document.createElement("td");

                            // Create a new anchor element
                            var link = document.createElement("a");

                            link.href = name + ".html";
                            link.textContent = name;

                            // Append the anchor element to the cell
                            cell.appendChild(link);

                            // Append the cell to the row
                            row.appendChild(cell);

                            // Append the row to the table body
                            tbody.appendChild(row);

                        });

                    </script>
                </tbody>
            </table>
        </body>
        </html>'''

    open_path = f'../webpages/{webpage_name}.html' # put a variable within a string
    f = open(open_path,'w')
    f.write(html_string)
    f.close()
```

```python
function_name_df = pd.read_csv("../functions_data/High_AUC_PPV_FUNCTIONS.txt", sep="\t")
function_name_df
```

```python
function_names_list = function_name_df["GO_LABEL"].tolist()
function_names_list = json.dumps(function_names_list)
function_names_list
```

```python
# function_names_string = ' '.join([str(name) for name in function_names_list])
# function_names_string
```

```python
webpage_name = "index"
convert_to_HTML_format(function_names_list, webpage_name)
```
